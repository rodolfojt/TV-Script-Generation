```python
# Treinando Dictionaries

import helper
import numpy as np
from collections import Counter

data_dir = './data/simpsons/moes_tavern_lines.txt'
text = helper.load_data(data_dir)

text = text[81:]

text = text.lower()
text = text.split()

# Cria um contador para todas palavras em caixa baixa do texto
counts = Counter(text)

# Retorna uma lista com o contador counts em ordem de frequência 
# do contador(key=counts.get) decrescente (reverse é True)
vocab = sorted(counts, key=counts.get, reverse= True)

# Cria um dictionaru que atribui um inteiro de acordo com a frequencia da palavra em vocab
vocab_to_int = {c: i for i, c in enumerate(vocab, 1)} # tirar dúvida dessa linha

# Lista que converterá o texto em inteiros
text_to_int = []

# Converte o texto em inteiros de acordo com o dicionário vocab_to_int
for each in text:
    text_to_int.append(vocab_to_int[each])

# Cria um dictionary que associa um inteiro a uma palavra do vocab
int_to_vocab = {i: c for i, c in enumerate(vocab, 1)}
```
